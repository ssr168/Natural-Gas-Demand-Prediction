```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df80_temp= pd.read_csv('data/all_temp/us_80_84.csv')
df85_temp= pd.read_csv('data/all_temp/us_85_89.csv')
df90_temp= pd.read_csv('data/all_temp/us_90_94.csv')
df95_temp= pd.read_csv('data/all_temp/us_95_99.csv')
df00_temp= pd.read_csv('data/all_temp/us_00_04.csv')
df05_temp= pd.read_csv('data/all_temp/us_05_09.csv')
df10_temp= pd.read_csv('data/all_temp/us_10_15.csv')
df15_temp= pd.read_csv('data/all_temp/us_15_19.csv')
```

```python
def df_cleaner(df):
    df['DateTime'] = df['DateTime'].map(lambda x:x[:-6])
    df['DateTime'] = pd.to_datetime(df['DateTime'])
    temp = df.columns[-1]
    df_clean = df.rename(index=str, columns={temp: "Temp"})
    return df_clean
```

```python
df80 = df_cleaner(df80_temp)
df85 = df_cleaner(df85_temp)
df90 = df_cleaner(df90_temp)
df95 = df_cleaner(df95_temp)
df00 = df_cleaner(df00_temp)
df05 = df_cleaner(df05_temp)
df10 = df_cleaner(df10_temp)
df15 = df_cleaner(df15_temp)
```

```python
dftemp = pd.concat([df80, df85, df90, df95, df00, df05, df10, df15])
dftemp.reset_index(drop=True)
```

```python
fig, ax = plt.subplots(figsize=(30,5))
ax.plot(dftemp['DateTime'], dftemp['Temp'])
```

```python
dftemp.to_csv(path_or_buf='data/temp_all_80_19.csv', index=False)
```
